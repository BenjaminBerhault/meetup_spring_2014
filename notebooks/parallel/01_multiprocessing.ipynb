# Multiprocessing

We are going to parallelize the [Mandelbrot set](http://en.wikipedia.org/wiki/Mandelbrot_set)

The Mandelbrot set is a mathematical set of points whose boundary is a distinctive and easily recognizable two-dimensional fractal shape. Images of the Mandelbrot set are made by sampling complex numbers and determining for each whether the result tends towards infinity when a particular mathematical operation is iterated on it. Treating the real and imaginary parts of each number as image coordinates, pixels are colored according to how rapidly the sequence diverges, if at all. --**wikipedia**

Reference: [Ian Ozsvald's tutorial](http://ianozsvald.com/2011/07/25/high-performance-python-tutorial-v0-2-from-europython-2011/)

## A few helper functions

```python
import numpy as np
np.seterr(all='ignore')
```

Sets up the grid of complex points.

```python
def grid(xvals, yvals):
    data = np.zeros(len(xvals)*len(yvals), np.complex128)
    index = 0
    for x in xvals:
        for y in yvals:
            data[index] = complex(x,y)
            index += 1
    return data
```

The computation for the  Mandelbrot.

```python
def mandel_z(q):
    
    z = q.copy()
    maxit = 256
    output = np.zeros(q.shape)
    
    for iteration in range(maxit):
        output += np.less(np.abs(z), 2.0)
        z = z*z + q
    return output
```

Creates the image result.

```python
def create_mandel(min_x, max_x, min_y, max_y, size):    
    
    x_vals = np.linspace(min_x, max_x, num=size)
    y_vals = np.linspace(min_y, max_y, num=size)
    
    mandelbrot = grid(x_vals, y_vals)
    results = mandel_z(mandelbrot)
    
    return results
```

Plots the result as an image.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import smash

def plot_mandel(z):
    Z = np.reshape(z, (size,size), order='F') 

    fig, ax = plt.subplots(figsize=(8,8))
    im = ax.imshow(Z, cmap=plt.cm.RdBu,  
            vmin=abs(Z).min(), 
            vmax=abs(Z).max(), 
            extent=[-2.2, 0.8, -1.5, 1.5])
    ax.grid(False)
```

### Break problem into blocks 

```python
def grid(xvals, yvals, block):
    index = 0
    data = []
    tmp = np.zeros(block, np.complex128)
    for x in xvals:
        for y in yvals:
            if (index)%block == 0:
                index = 0
                data.append(tmp)
                tmp = np.zeros(block, np.complex128)
            tmp[index] = complex(x,y)
            index+=1
    data.append(tmp)
    tmp = data.pop(0)
    return data
```

Use the `map` function.  Pass `mandel_z` as a parameter.

```python
def create_mandel(min_x, max_x, min_y, max_y, size, block_size):    
    x_vals = np.linspace(min_x, max_x, num=size)
    y_vals = np.linspace(min_y, max_y, num=size)
    
    mandelbrot = grid(x_vals, y_vals, block_size)

    results = map(mandel_z, mandelbrot)
    
    output = np.concatenate(results)
    
    return output
```

## Parallel, in 3 easy steps

Step 1: `import`

```python
import multiprocessing
```

Step 2: Create a `pool` of workers.

```python
num_cores = multiprocessing.cpu_count()
pool = multiprocessing.Pool(num_cores)
print num_cores
```

Step 3: Modify the serial `map`.

```python
def create_mandel(min_x, max_x, min_y, max_y, size, block_size, pool):    
    x_vals = np.linspace(min_x, max_x, num=size)
    y_vals = np.linspace(min_y, max_y, num=size)
    
    mandelbrot = grid(x_vals, y_vals, block_size)
    
    results = pool.map(mandel_z, mandelbrot)  # <- Modification
    
    output = np.concatenate(results)
    
    return output
```

Enjoy!

```python
%time z = create_mandel(-2.2,0.8,-1.5,1.5, size, block_size, pool)
```

## Essential Metrics

How efficient is this process?

```python
t = []
cpus = [1,2,4,6,8,10,12]
for cpu in cpus:
    pool = multiprocessing.Pool(cpu)
    tic = time.time()
    z = create_mandel(-2.2,0.8,-1.5,1.5,size, block_size, pool)
    toc = time.time() - tic
    t.append(toc)
    print cpu, toc
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(cpus, t,'-o')
ax.set_xlabel('number of cores')
ax.set_ylabel('time (sec)')
fig.show()
```

## Speedup and Efficiency

    speedup = sequential time / parallel time

\begin{eqnarray*}
s &=& \frac{t_1}{t_n} \\
\end{eqnarray*}

Measures processor utilization

    efficiency = sequential time / ( number of processors ⋅ parallel time )
               = speedup / numer of processors

\begin{eqnarray*}
e &=& \frac{s}{n}
\end{eqnarray*}

```python
import pandas as pd
df = pd.DataFrame({'n': cpus, 't': t})
```

```python
df['speedup'] = df[df['n'] == 1]['t'].values/df['t']
df['eff'] = df['speedup']/df['n']
```

### Speedup

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(df['n'], df['speedup'].values, label='mandel')
ax.plot(range(12), range(12), label='ideal')
ax.set_xlabel('number of cores')
ax.set_ylabel('speedup')
ax.legend(loc='upper left')
fig.show()
```

### Efficiency

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(df['n'], df['eff'].values)
ax.set_xlabel('number of cores')
ax.set_ylabel('efficiency')
ax.set_ylim(0,1.)
fig.show()
```

### Amdahl's Law

```python
parallel_ = 0.9
serial_ = 0.1
total_time = df.ix[0,'t']

np = range(1, 13)
t =  [ total_time * (serial_ + parallel_/float(n)) for n in np ]
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(np, total_time/t, label='Amdahl')
ax.plot(df['n'], df['speedup'].values, label='actual')
ax.set_xlabel('number of cores')
ax.set_ylabel('speedup')
ax.legend(loc='upper left')
fig.show()
```
