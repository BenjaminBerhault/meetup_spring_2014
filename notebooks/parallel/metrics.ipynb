```python
import numpy as np
np.seterr(all='ignore')
```

```python
def grid(xvals, yvals):
    data = np.zeros(len(xvals)*len(yvals), np.complex128)
    index = 0
    for x in xvals:
        for y in yvals:
            data[index] = complex(x,y)
            index += 1
    return data

def mandel_z(q):
    
    z = q.copy()
    maxit = 256
    output = np.zeros(q.shape)
    
    for iteration in range(maxit):
        output += np.less(np.abs(z), 2.0)
        z = z*z + q
    return output

def create_mandel(min_x, max_x, min_y, max_y, size):    
    
    x_vals = np.linspace(min_x, max_x, num=size)
    y_vals = np.linspace(min_y, max_y, num=size)
    
    mandelbrot = grid(x_vals, y_vals)
    results = mandel_z(mandelbrot)
    
    return results
```

```python
size = 1000
%time z = create_mandel(-2.2, 0.8, -1.5, 1.5, size)
```

```python
def grid(xvals, yvals, block):
    index = 0
    data = []
    tmp = np.zeros(block, np.complex128)
    for x in xvals:
        for y in yvals:
            if (index)%block == 0:
                index = 0
                data.append(tmp)
                tmp = np.zeros(block, np.complex128)
            tmp[index] = complex(x,y)
            index+=1
    data.append(tmp)
    tmp = data.pop(0)
    return data
```

```python
def create_mandel(min_x, max_x, min_y, max_y, size, block_size):    
    x_vals = np.linspace(min_x, max_x, num=size)
    y_vals = np.linspace(min_y, max_y, num=size)
    
    mandelbrot = grid(x_vals, y_vals, block_size)
    
    results = map(mandel_z, mandelbrot)
    
    output = np.concatenate(results)
    
    return output
```

```python
import multiprocessing
num_cores = multiprocessing.cpu_count()
pool = multiprocessing.Pool(num_cores)
print num_cores
```

```python
def create_mandel(min_x, max_x, min_y, max_y, size, block_size, pool):    
    x_vals = np.linspace(min_x, max_x, num=size)
    y_vals = np.linspace(min_y, max_y, num=size)
    
    mandelbrot = grid(x_vals, y_vals, block_size)
    
    results = pool.map(mandel_z, mandelbrot)
    
    output = np.concatenate(results)
    
    return output
```

```python
%time z = create_mandel(-2.2,0.8,-1.5,1.5,size,block_size, pool)
```

```python
t = []
cpus = [1,2,4,6,8,10,12]
for cpu in cpus:
    pool = multiprocessing.Pool(cpu)
    tic = time.time()
    z = create_mandel(-2.2,0.8,-1.5,1.5,size,block_size, pool)
    t.append(time.time() - tic)
```

```python
plt.plot(cpus, t,'-o')
plt.show()
```

## Speedup and Efficiency

    speedup = sequential time / parallel time

\begin{eqnarray*}
s &=& \frac{t_1}{t_n} \\
\end{eqnarray*}

Measures processor utilization

    efficiency = sequential time / ( number of processors ⋅ parallel time )
               = speedup / numer of processors

\begin{eqnarray*}
e &=& \frac{s}{n}
\end{eqnarray*}

```python
import pandas as pd
df = pd.DataFrame({'n': cpus, 't': t})
```

```python
df['speedup'] = df[df['n'] == 1]['t'].values/df['t']
df['eff'] = df['speedup']/df['n']
```

```python
plt.plot(df['n'], df['speedup'].values, '-o')
plt.plot(range(12), range(12))
plt.show()
```

```python
plt.plot(df['n'], df['eff'].values, '-o')
plt.show()
```

```python
plt.plot(np, total_time/t, '-o')
plt.plot(df['n'], df['speedup'].values, '-o')
plt.show()
```
